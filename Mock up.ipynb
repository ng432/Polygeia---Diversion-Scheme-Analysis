```python
import numpy as np
import geopandas as gp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# setting up police force county data frames

Count = gp.read_file('UKcountries')
Count = Count.drop(columns = ['objectid', 'ctry18cd', 'ctry18nmw', 'bng_e', 'bng_n', 'long', 'lat', 'st_areasha', 'st_lengths']) 
Count = Count.rename(columns={'ctry18nm': 'name'})

PA = gp.read_file('PA')
PA = PA.drop(columns = ['OBJECTID', 'PFA16CD', 'BNG_E', 'BNG_N', 'LONG', 'LAT', 'Shape__Are', 'Shape__Len'])
PA = PA.rename(columns={'PFA16NM': 'name'})

frames = [Count, PA]
all_PA = pd.concat(frames)
                                       
```

```python


# diversion scheme code
def add_fake_data(data_frame):

    example_info_num = np.random.randint(3, size=len(data_frame.name))
    example_info_t = [None] * len(data_frame.name)

    j = 0

    # 0, no Diversion scheme, 1 planned, 2 existed
    for i in example_info_num:

        if i == 0:
            example_info_t[j] = 'No diversion scheme'
        elif i == 1: 
            example_info_t[j] = 'Diversion scheme planned'
        elif i == 2: 
            example_info_t[j] = 'Diversion scheme in place'

        j +=1

    
    data_frame['Diversion Scheme Status'] = example_info_t
       
    #data_frame.map()
    
    return data_frame
```

```python
all_PA = add_fake_data(all_PA)

```

```python
colour_dict = 'summer_r'

fig, ax = plt.subplots(1, figsize=(10,10))

all_PA.plot('Diversion Scheme Status', ax=ax, legend=True, legend_kwds={'fontsize': 10, 'loc': 'upper left'}, cmap=colour_dict)
plt.axis('off')
ax.set_title('Diversion scheme implementation across UK police forces')

plt.savefig('Figures/quick_mock_up.pdf', bbox_inches='tight')
```
